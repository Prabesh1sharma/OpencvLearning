```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

`Image Histogram : ` It  is plot with pixel values(ranging from 0 to 255, not always) in X-axis and corresponding number of pixels in the image on Y-axis

```python
blank_w = np.ones((500,500,3),np.uint8)*255
blank_b = np.zeros((500,500,3),np.uint8)*255
```

```python
hist = cv2.calcHist([blank_w],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
cv2.imshow("blank w",blank_w)
# cv2.imshow("blank b",blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist = cv2.calcHist([blank_b],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
cv2.imshow("blank w",blank_b)
# cv2.imshow("blank b",blank_w)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Histogram Equalization in opencv

- Histogram equalization is a method in image processing of contrast adjustment using image's histogram.
- This method usually increases the global contrast of many images, especially when the usable data of the image is represented by close contrast values.
### CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
img = cv2.imread("Images/hill.jpeg")
img = cv2.resize(img,(800,700))
gry =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist1 = cv2.calcHist([gry],[0],None,[256],[0,256])
eq = cv2.equalizeHist(gry)
hist2 = cv2.calcHist([eq],[0],None,[256],[0,256])
plt.plot(hist1)
plt.plot(hist2,c='red')
plt.show()
cv2.imshow("blank r",img)
cv2.imshow("prabesh",eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("Images/hill.jpeg")
img = cv2.resize(img,(800,700))
gry =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist1 = cv2.calcHist([gry],[0],None,[256],[0,256])
eq = cv2.equalizeHist(gry)
hist2 = cv2.calcHist([eq],[0],None,[256],[0,256])

cl = cv2.createCLAHE(clipLimit = 1, tileGridSize = (8,8))
f = cl.apply(gry)
hist3 = cv2.calcHist([f],[0],None,[256],[0,256])
plt.plot(hist1)
plt.plot(hist2,c='red')
plt.plot(hist3,c='green')
plt.show()
cv2.imshow('clay',f)
cv2.imshow("blank r",img)
cv2.imshow("gray",eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Mouse event Bindings using Opencv python

```python
import cv2
import numpy as np

# Define the event handling function
def handle_mouse_event(event, x, y, flags, param):
    global img
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 5, (0, 0, 255), 4)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img, (x, y), (x + 20, y + 20), (255, 0, 0), 4)

# Create a named window
cv2.namedWindow('Hancy')

# Initialize a white image
img = np.ones((600, 600, 3), np.uint8) * 255

# Set the mouse callback function to the window
cv2.setMouseCallback('Hancy', handle_mouse_event)

# Display the window and handle key events
while True:
    cv2.imshow('Hancy', img)
    if cv2.waitKey(25) & 0xFF == ord('p'):
        break

# Destroy all OpenCV windows
cv2.destroyAllWindows()
```

# Contour Moments and Convex Hull using Opencv

```python
import cv2
import numpy as np

# Read and resize the image
img = cv2.imread("Images/shapes.webp")
img = cv2.resize(img, (400, 400))

# Convert the image to grayscale
gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, thr = cv2.threshold(gry, 210, 255, cv2.THRESH_BINARY)

# Find contours
cnt, htr = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

ar = []
# Iterate through each contour
for c in cnt:
    m = cv2.moments(c)
    if m['m00'] != 0:
        x = int(m["m10"] / m['m00'])
        y = int(m["m01"] / m['m00'])
        cv2.circle(img, (x, y), 2, (255, 0, 0), -1)  # Draw the center of the contour
    cv2.drawContours(img, [c], -1, (0, 0, 255), 2)
    a = cv2.contourArea(c)
    ar.append(a)
    ep = 0.1*cv2.arcLength(c,True) #decrease the value of 0.01 value to cover all the boundaries
    d = cv2.approxPolyDP(c,ep,True)
    h = cv2.convexHull(d)
    x,y,w,h = cv2.boundingRect(h)
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)
print(ar)
# Display the results
cv2.imshow("opencvthr", thr)
cv2.imshow("opencv", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
